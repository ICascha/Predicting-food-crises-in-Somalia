```python
import matplotlib.pyplot as plt
```

```python
def nickell_bias(T, gamma):
    part1 =  ((1 + gamma)/(T - 1)) * (1 - (1/T)*(1-gamma**T)/(1 - gamma))
    part2 = (1 - (2 * gamma/((1 - gamma)*(T - 1))) *  (1 - (1/T)*(1-gamma**T)/(1 - gamma)))**(-1)
    
    return -1 * part1 * part2
```

```python
fig, ax = plt.subplots()

T_range = list(range(2, 101))

ax.plot(T_range, [nickell_bias(T, 0.1) for T in T_range])
ax.plot(T_range, [nickell_bias(T, 0.5) for T in T_range])
ax.plot(T_range, [nickell_bias(T, 0.9) for T in T_range])
ax.set_xlabel('T')
ax.set_ylabel('Nickell bias')
ax.set_xticks(range(0, 110, 10))
ax.set_xlim(0, 100)
ax.set_ylim(-0.5, 0)
ax.grid(alpha=0.7)
ax.legend(['$\\gamma = 0.1$', '$\\gamma = 0.5$', '$\\gamma = 0.9$'])

fig.savefig('../figures/nickell_bias.pdf', format='pdf')
```
